```python
from sklearn import datasets
iris = datasets.load_iris()

list(iris.keys())
```

```python
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
print(iris.data)
```

```python
print(iris.target_names)
```

```python
print(iris.target)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = (iris["target"] == 2).astype(np.int)
X = iris["data"][:,2:]
```

```python
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
xx, yy = np.meshgrid(np.arange(0,8, 0.01),np.arange(0, 3.0, 0.01))
Z = log_reg.predict(np.c_[xx.ravel(),yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.figure(1, figsize = (8, 6))
plt.pcolormesh(xx, yy, Z, cmap = "RdBu", vmin = -.2, vmax = 1.2)
plt.scatter(X[:,0],X[:,1], c=y, s=50,
            cmap="RdBu",vmin=-.2,vmax=1.2,
            edgecolor="white",linewidth=1)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
```

```python
np.sum(log_reg.predict(X)!=y)
```

#두 번째 영상

```python
y = iris.target
X=iris["data"][:,2:]
```

```python
softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10)
softmax_reg.fit(X,y)
```

```python
xx, yy = np.meshgrid(np.arange(0,8,0.01),np.arange(0,3.0,0.01))
Z = softmax_reg.predict(np.c_[xx.ravel(),yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.figure(1,figsize = (8,6))
plt.pcolormesh(xx,yy,Z,cmap="RdBu",vmin=-0.2,vmax=1.2)
plt.scatter(X[:,0],X[:,1],c=y,cmap="RdBu",
            vmin=-.2,vmax=1.2,
            edgecolor="white",linewidth = 1)

plt.xlabel("Petal length")
plt.ylabel("Petal width")
```

```python
np.sum(softmax_reg.predict(X)!=y)
```

Multinomial regression using 4 features & MLPClassifier

```python
y = iris.target
X=iris["data"]
```

```python
from sklearn.neural_network import MLPClassifier

classifier_MLP = MLPClassifier(activation = 'relu', hidden_layer_sizes = (16,32,16), max_iter = 20000, solver = "lbfgs")
classifier_MLP.fit(X,y)
```

```python
np.sum(classifier_MLP.predict(X)!=y)
```
